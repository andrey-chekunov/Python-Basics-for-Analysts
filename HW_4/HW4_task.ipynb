### Условие 2: 2 задача
#### 2.1 Изучите распределение домов от наличия вида на набережную
Постройте график
Сделайте выводы

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

#### 2.2 Изучите распределение этажей домов

```python
import seaborn as sns

plt.figure(figsize=(14, 8))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='Spectral');
```

#### 2.2 Изучите распределение состояния домов

```python
import seaborn as sns

plt.figure(figsize=(14, 8))
sns.heatmap(corr_matrix, annot = True, cbar_kws= {'orientation': 'horizontal'} )
```

```python
plt.figure(figsize=(8, 12))
heatmap = sns.heatmap(corr_matrix.corr()[['sqft_living']].sort_values(by='sqft_living', ascending=False), vmin=-1, vmax=1, annot=True, cmap='BrBG')

heatmap.set_title('Распределение по состоянию домов', fontdict={'fontsize':18}, pad=16);


```

```python
plt.figure(figsize=(8, 6))

sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg')
plt.title('Отношение цены и площади недвижимости');
```

```python
plt.figure(figsize=(8, 6))
sns.jointplot(x=df['price'], y=df['bedrooms'], kind='reg')
plt.title('Отношение цены и количества спален');
```

Вывод:

Количество спален не критично влияет на стоимость недвижимости

```python
plt.figure(figsize=(8, 6))
sns.jointplot(x=df['price'], y=df['bathrooms'], kind='reg')
plt.title('Отношение цены и ванных комнат');
```

Вывод:

Ванные комнаты не влияют на стоимость
